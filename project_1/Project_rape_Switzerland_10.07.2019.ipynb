```python
df.loc[df.exact_punish.str.contains('Jahr', na=False), 'cleaned_number_in_months'] = df.number_punish.mul(12)
df.loc[df.exact_punish.str.contains('Tag', na=False), 'cleaned_number_in_months'] = df.number_punish.div(30)
df.loc[df.exact_punish.str.contains('Monat', na=False), 'cleaned_number_in_months'] = df.number_punish
df.head()
```

```python
df.loc[df.exact_punish1.str.contains('Jahr', na=False), 'cleaned_number_in_months1'] = df.number_punish1.mul(12)
df.loc[df.exact_punish1.str.contains('Tag', na=False), 'cleaned_number_in_months1'] = df.number_punish1.div(30)
df.loc[df.exact_punish1.str.contains('Monat', na=False), 'cleaned_number_in_months1'] = df.number_punish1
df.head()
```

```python
df['year'] = df.year.str.extract("(^\d\d\d\d)")
```

```python
df1 = df[['year','title', 'cleaned_number_in_months', 'cleaned_number_in_months1']]
df1['months'] = df1['cleaned_number_in_months']
df1['months1'] = df1['cleaned_number_in_months1']
df1.head()
df1 = df1[['year','title', 'months', 'months1']]
df1.head()
```

```python
df1.months[(df1.months == 'Nan')] = df1['months1']
df1.head()
```
