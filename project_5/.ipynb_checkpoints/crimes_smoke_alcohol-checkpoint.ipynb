```python
df = pd.read_csv('tabula-Alkohol_Konsum_all.csv')
```

```python
df['Gesamtkonsum_Rein_Alk'] = df['Gesamtkonsum_Rein_Alk'].str.replace('9 .2', '9.2')
```

```python
df['Gesamtkonsum_Rein_Alk'] = df['Gesamtkonsum_Rein_Alk'].str.replace('8 B.7', '8.7')
```

```python
df['Gesamtkonsum_Rein_Alk'] = df['Gesamtkonsum_Rein_Alk'].str.replace('8S.5', '8.5')
```

```python
df['Gesamtkonsum_Rein_Alk'] = df['Gesamtkonsum_Rein_Alk'].str.replace('8 G.1', '8.1')
```

```python
df['Gesamtkonsum_Rein_Alk'] = df['Gesamtkonsum_Rein_Alk'].str.replace('ko\r8.1', '8.1')
```

```python
df['Gesamtkonsum_Rein_Alk'] = df['Gesamtkonsum_Rein_Alk'].replace(np.nan, '7.7')
```
