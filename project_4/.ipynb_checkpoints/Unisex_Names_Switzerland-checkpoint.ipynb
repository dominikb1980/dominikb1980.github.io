```python
# Example Python program to draw a percentage

# based area plot(stacked) for a Python pandas DataFrame

 

import pandas as pd

import matplotlib.pyplot as plot


values = {"Girls"   : list(merged.loc[merged.Vorname == 'Lou', merged.columns.str.endswith("x")].sort_index(axis=1).T.iloc[:,0].values),
          "Boys"  : list(merged.loc[merged.Vorname == 'Lou', merged.columns.str.endswith("y")].sort_index(axis=1).T.iloc[:,0].values)};


dataFrame           = pd.DataFrame(data=values);
# Normalize Data

normalizedDataFrame = dataFrame.div(dataFrame.sum(axis=1), axis=0) *100;
#years = ("1998", "1999", "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017");
years =list(range(1998, 2018, 1))
normalizedDataFrame.index = years;
pal = ["#f47f7b", "#008ca6"]
# Draw a percentage based, stacked area plot

ax = normalizedDataFrame.plot.area(stacked=True, color=pal, legend=False);

# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)

# plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

plot.title('Lou', loc='left', fontsize=35, fontweight="bold")
plot.box(False)
plot.yticks(range(0, 101, 50), fontsize=25)
plot.xticks(range(2000, 2018, 8), fontsize=25)
ax.axes.get_yaxis().set_visible(False)
plot.savefig('new_lou.pdf')
plot.show(block=True)
```

```python
def name_plot(name):
    values = {"Girls"   : list(merged.loc[merged.Vorname == name, merged.columns.str.endswith("x")].sort_index(axis=1).T.iloc[:,0].values),
          "Boys"  : list(merged.loc[merged.Vorname == name, merged.columns.str.endswith("y")].sort_index(axis=1).T.iloc[:,0].values)};

    dataFrame= pd.DataFrame(data=values);
    # Normalize Data

    normalizedDataFrame = dataFrame.div(dataFrame.sum(axis=1), axis=0) *100;
    years= ("1998", "1999", "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017");

    normalizedDataFrame.index = years;
    pal = ["#f47f7b", "#008ca6"]
    # Draw a percentage based, stacked area plot

    normalizedDataFrame.plot.area(stacked=True, colors=pal);

    plot.title(name, loc='left', fontsize=35, fontweight="bold")
    plot.box(False)
    plot.show(block=True)
```

```python
def name_plot(name):
    values = {"Girls"   : list(merged.loc[merged.Vorname == name, merged.columns.str.endswith("x")].sort_index(axis=1).T.iloc[:,0].values),
          "Boys"  : list(merged.loc[merged.Vorname == name, merged.columns.str.endswith("y")].sort_index(axis=1).T.iloc[:,0].values)};


    dataFrame           = pd.DataFrame(data=values);
    # Normalize Data

    normalizedDataFrame = dataFrame.div(dataFrame.sum(axis=1), axis=0) *100;
    #years = ("1998", "1999", "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017");
    years =list(range(1998, 2018, 1))
    normalizedDataFrame.index = years;
    pal = ["#f47f7b", "#008ca6"]
    # Draw a percentage based, stacked area plot

    ax = normalizedDataFrame.plot.area(stacked=True, color=pal, legend=False);

    # ax.spines['top'].set_visible(False)
    # ax.spines['right'].set_visible(False)

    # plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

    plot.title(name, loc='left', fontsize=35, fontweight="bold")
    plot.box(False)
    plot.yticks(range(0, 101, 50), fontsize=25)
    plot.xticks(range(2000, 2018, 8), fontsize=25)
    ax.axes.get_yaxis().set_visible(False)
    # plot.axis('off')
    # plot.show(block=True)
    plot.savefig(name + '_.pdf')
    plot.show(block=True)
```

```python
for name in list(merged.Vorname.values):
    name_plot(name)
```
